# Minimal self-supervised conformal prediction demo

This notebook demonstrates the public API on a tiny Gaussian identity-denoising problem. It is not a paper benchmark.

```python
import torch
from torch.utils.data import DataLoader
from sscp.data.synthetic import GaussianIdentityDataset
from sscp.conformal.calibration import calibrate_self_supervised
from sscp.conformal.coverage import empirical_coverage
from sscp.conformal.scores import make_measurement_estimator
from sscp.losses.sure import gaussian_sure
```

```python
sigma = 0.1
calibration = GaussianIdentityDataset(num_samples=64, sigma=sigma, seed=0)
test = GaussianIdentityDataset(num_samples=32, sigma=sigma, seed=1)
estimator = lambda y: 0.8 * y
h_fn = make_measurement_estimator(estimator)
risk = lambda y: gaussian_sure(y, h_fn=h_fn, sigma=sigma)
```

```python
result = calibrate_self_supervised(DataLoader(calibration, batch_size=8), risk, confidence=0.9)
coverage = empirical_coverage(DataLoader(test, batch_size=8), estimator, None, result.q_hat)
float(result.q_hat), coverage
```
